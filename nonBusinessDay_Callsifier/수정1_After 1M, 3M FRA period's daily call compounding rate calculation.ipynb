### **daily_compounding_rate2**
- type이 pandas timestamp 이라면 to_pydatetime() method 활용해 datetime으로 수정

https://velog.io/@jaehwi0823/pandas-Timestamp%EC%9D%84-python-datetime%EC%9C%BC%EB%A1%9C-%EB%B3%80%ED%99%9

```python
def daily_compounding_rate2(starting_date, ending_date):
    
    '''
    start_date는 콜을 빌리는 오후 일자 / string 객체인 매개변수
    end_date는 콜을 갚는 오후 일자  / string 객체인 매개변수

    i.e. start_date의 콜금리는 포함되어야 하지만, end_date의 콜금리는 포함되면 안된다
    '''  
    
    try: 
        '''type이 pandas timestamp 이라면 to_pydatetime() method 활용해 datetime으로 수정'''
        
        starting_date = starting_date.to_pydatetime()
        ending_date = ending_date.to_pydatetime()
        
    except: 
        '''원래 datetime이라면 그대로 적용'''
        
        starting_date = starting_date
        ending_date = ending_date
    
    term = ending_date - starting_date 
    
    
    # matche call_close with date
    starting_date_index = df.index[(df['call_date'] == starting_date)]
    ending_date_index = df.index[(df['call_date'] == ending_date)]
    
    
    # define "daily_call_rate_list"
    daily_call_rate_list = []
    
    for i in range(term.days): 
        daily_call_rate_list.append(df['call_close'][starting_date_index-i].values[0])
     
    
    # define "compounding_rate"
    compound_rate = 1
    
    for i in range(term.days):
        compound_rate = compound_rate * (1 + (daily_call_rate_list[i]/100/365))   # daily compound
                       
    return (compound_rate - 1)*100 

# 최종적으로 해당 함수반환값이 곧 ff(t+j, t+j+k) 값과 동일하다!!!!
# 함수 반환값은 무조건 "해당기간의 복리 수익률(%)"이라는점 반드시 명심하고, 다른 ff(t), r(t)도 수익률(%)로 맞춰주기!  
```

### FRA generate datetime

```python
# argument is not a current row's Date but after 1M & after 1M + 3M 

for idx in df2.index:
    
    # generate datetime object for FRA sheet
    df2.loc[idx, 'Date2'] = pd.to_datetime(df2.loc[idx, 'Date'])
    
    # define after 1M as after 30days
    df2.loc[idx, "1M_after_start_date"] = df2.loc[idx, 'Date2'] + timedelta(days=30)

    # typical FRA means after 91 days
    df2.loc[idx, "1M_FRA_mature_date"] = df2.loc[idx, '1M_after_start_date'] + timedelta(days=91)
    
df2.head()
```

```python
df2.info()
```

```python
# calculate 1day call compounding rate for 1M after 3M FRA period - daily_compounding_rate2 적용
for idx in df2.index:
    df2.loc[idx, "After_01M_3Mcall"] = daily_compounding_rate2(df2.loc[idx, '1M_after_start_date'], df2.loc[idx, '1M_FRA_mature_date'])
```

```python
df2.info()
```

```python
# generate new DataFrame to be joined.
df3 = df[["call_date","call_close"]]
df4 = df2[["Date2","01M*04M_MID","After_01M_3Mcall"]]

# merge(inner join)기능 수행하기 위해 df3, df4 날짜(call_date, Date2)열 이름 동일하게 맞춘 상태로 merge하기.
df5 = pd.merge(df3.rename(columns={'call_date':'Date2'}), df4)
```
