```python
def return_name(market):
    '''get name from ticker'''
    Market = []
    for ticker in market:
        Value = stock.get_market_ticker_name(ticker)
        Market.append([Value,ticker])
 
    df = pd.DataFrame(Market,columns=['회사명','상장번호'])
    return df

KOSPI = stock.get_market_ticker_list(market="KOSPI")
kospi = return_name(KOSPI)
```

코스피와 intercept 칼럼 넣기

```python
kospi_data = web.DataReader('^KS11',
                             start="2016-01-01",
                             end='2021-12-31', actions=False)
kospi_ret = kospi_data['Close'].pct_change()
kospi_ret # 코스피 데이터
```

```python
prices = pd.DataFrame({tic: data['Close'] for tic, data in all_data.items()})
```

```python
ret = prices.dropna(axis=1).pct_change().dropna()
ret['^KS11'] = kospi_ret
ret['intercept'] = 1 # 코스피와 intercept 칼럼 넣기
```

```python
ret_col = ret.columns[:-2] # 개별 종목 ticker 까지 columns설정
ret_col
```

```python
beta_data = {}

for ticker in ret_col:
    reg = sm.OLS(ret[[ticker]], ret[['^KS11', 'intercept']]).fit()
    beta_data[ticker] = reg.params[0]
```

```python
beta_data
beta_series=pd.Series(beta_data)
beta_series.sort_values()
```

```python
beta_bottom20= beta_series.sort_values()[:20]
beta_top20 = beta_series.sort_values()[-20:]
```

```python
beta_bottom20.index
beta_bottom20.values

ticker_name = []
for i in beta_bottom20.index:
    # print(i.split('.')[0])
    # print(stock.get_market_ticker_name(i.split('.')[0]))
    ticker_name.append(stock.get_market_ticker_name(i.split('.')[0]))

pd.DataFrame({'ticker': beta_bottom20.index,
             'name': ticker_name,
            'beta': beta_bottom20.values})
```

 통신업과 섬유ㆍ의복, 의료정밀, 제약, 종이ㆍ목재, 비금속, 전기ㆍ가스 등이 저베타업종으로 분류된다

```python
ticker_name = []
for i in beta_top20.index:
    # print(i.split('.')[0])
    # print(stock.get_market_ticker_name(i.split('.')[0]))
    ticker_name.append(stock.get_market_ticker_name(i.split('.')[0]))

pd.DataFrame({'ticker': beta_top20.index,
             'name': ticker_name,
            'beta': beta_top20.values}).sort_values(by='beta',ascending=False)
```

### 분위별 시각화 함수
- price_data 추가

```python
def quantile_plot(df, col, price_data):
    q10=df[col].quantile(q=1, interpolation='nearest')
    q8 =df[col].quantile(q=0.8, interpolation='nearest')
    q6 =df[col].quantile(q=0.6, interpolation='nearest')
    q4 =df[col].quantile(q=0.4, interpolation='nearest')
    q2 =df[col].quantile(q=0.2, interpolation='nearest')

    q10_8 = df.iloc[df[df[col] == q10].index[0]:df[df[col] == q8].index[0],:]
    q8_6 = df.iloc[df[df[col] == q8].index[0]:df[df[col] == q6].index[0],:]
    q6_4 = df.iloc[df[df[col] == q6].index[0]:df[df[col] == q4].index[0],:]
    q4_2 = df.iloc[df[df[col] == q4].index[0]:df[df[col] == q2].index[0],:]
    q2_0 = df.iloc[df[df[col] == q2].index[0]:,:]

    q10_8_ticker_list = [i for i in q10_8['종목코드']]
    q10_8_ret_sum = price_data[q10_8_ticker_list].sum(axis=1)
    
    q8_6_ticker_list = [i for i in q8_6['종목코드']]
    q8_6_ret_sum = price_data[q8_6_ticker_list].sum(axis=1)

    q6_4_ticker_list = [i for i in q6_4['종목코드']]
    q6_4_ret_sum = price_data[q6_4_ticker_list].sum(axis=1)

    q4_2_ticker_list = [i for i in q4_2['종목코드']]
    q4_2_ret_sum = price_data[q4_2_ticker_list].sum(axis=1)
    
    q2_0_ticker_list = [i for i in q2_0['종목코드']]
    q2_0_ret_sum = price_data[q2_0_ticker_list].sum(axis=1)
    
    df_last = pd.DataFrame({'q10_8':q10_8_ret_sum, 'q8_6':q8_6_ret_sum, 
                             'q6_4':q6_4_ret_sum, 'q4_2':q4_2_ret_sum, 
                             'q2_0':q2_0_ret_sum})

    df_last.plot(legend=True, figsize=(20,10), title =f'{col}');
    plt.grid(True, color = '0.7', linestyle =':', linewidth=1)
    plt.show();
```

```python
quantile_plot(df_beta_sorted,'beta',price_cumsum)
```
