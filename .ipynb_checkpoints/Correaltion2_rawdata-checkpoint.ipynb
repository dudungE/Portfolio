```python
import pandas as pd
from scipy import stats
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
import matplotlib.pyplot as plt
# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
ss = pdr.get_data_yahoo('005930.KS',start='2005-01-04')
msft = pdr.get_data_yahoo('MSFT',start='2005-01-04')

dow = pdr.get_data_yahoo('^DJI', '2005-01-04')
kospi = pdr.get_data_yahoo('^KS11', '2005-01-04')
```

```python
plt.figure(figsize=(15,10))

plt.subplot(211)
plt.plot(ss.index, ss.Close, 'b', label='Samsung')
plt.plot(msft.index, msft.Close, 'r--', label='Microsoft')
plt.grid(True)
plt.legend(loc='best');

plt.subplot(212)
plt.plot(dow.index, dow.Close, 'r--', label='dow')
plt.plot(kospi.index, kospi.Close, 'b', label='kospi')
plt.grid(True)
plt.legend(loc='best')
            
plt.show();
```

```python
kospi.info()
```

# 누적수익률

```python
ss_ret_daily = ss.pct_change().cumsum()
msft_ret_daily= msft.pct_change().cumsum()
dow_ret_daily.pct_change().cumsum()
kospi_ret_daily.pct_change().cumsum()

plt.figure(figsize=(15,10))

plt.subplot(211)
plt.plot(ss_ret_daily.index, ss_ret_daily.Close, 'b', label='Samsung')
plt.plot(msft_ret_daily.index, msft_ret_daily.Close, 'r--', label='Microsoft')
plt.grid(True)
plt.legend(loc='best');

plt.subplot(212)
plt.plot(kospi_ret_daily.index, kospi_ret_daily.Close, 'b', label='kospi')
plt.plot(dow_ret_daily.index, dow_ret_daily.Close, 'r--', label='dow')

plt.grid(True)
plt.legend(loc='best')
            
plt.show();
```

# 상관관계

```python
df_stock = pd.DataFrame({'SAMSUNG':ss['Close'], 'MSFT':msft['Close']})
df_stock.info()
print('========================================')
df_index = pd.DataFrame({'DOW':dow['Close'], 'KOSPI':kospi['Close']})
df_index.info()
```

```python
df_stock = df_stock.fillna(method='bfill')
df_stock = df_stock.fillna(method='ffill')
df_stock.info() # 결측치 채워주기
print('========================================')
df_index = df_index.fillna(method='bfill')
df_index = df_index.fillna(method='ffill')
df_index.info() # 결측치 채워주기
```

```python
regr_stock = stats.linregress(df_stock['MSFT'], df_stock['SAMSUNG'])
regr_stock_line = f'Y = {regr_stock.slope:2f}X + {regr_stock.intercept:2f}'
print(regr_stock)

regr_index = stats.linregress(df_index['DOW'],df_index['KOSPI'])
regr_index_line = f'Y = {regr_index.slope:2f}X + {regr_index.intercept:2f}'
print(regr_index)
```

```python
regr_stock_line
```

```python
regr_stock
```

```python
plt.figure(figsize=(15,8))

plt.subplot(121)
plt.plot(df_stock.MSFT, df_stock.SAMSUNG, 'x') 
plt.plot(df_stock.MSFT, regr_stock.slope * df_stock.MSFT + regr_stock.intercept, 'r')
plt.legend(['SAMSUNG x MSFT', regr_stock_line])
plt.title(f'SAMSUNG x MSFT (R_sqaured = {regr_stock.rvalue:2f})')
plt.xlabel('SAMSUNG')
plt.ylabel('MSFT')

plt.subplot(122)
plt.plot(df_index.DOW, df_index.KOSPI, 'x') 
plt.plot(df_index.DOW, regr_index.slope * df_index.DOW + regr_index.intercept, 'r')
plt.legend(['DOW x KOSPI', regr_index_line])
plt.title(f'DOW x KOSPI (R_sqaured = {regr_index.rvalue:2f})')
plt.xlabel('DOW')
plt.ylabel('KOSPI')
plt.show();
```

```python
plt.figure(figsize=(15,8))

plt.subplot(121)
plt.plot(df_stock.MSFT, df_stock.SAMSUNG, 'x') 
plt.plot(df_stock.MSFT, regr_stock.slope * df_stock.MSFT + regr_stock.intercept, 'r')
plt.legend(['SAMSUNG x MSFT', regr_stock_line])
plt.title(f'SAMSUNG x MSFT (R_sqaured = {regr_stock.rvalue:2f})')
plt.xlabel('SAMSUNG')
plt.ylabel('MSFT')

plt.subplot(122)
plt.plot(df_index.DOW, df_index.KOSPI, 'x') 
plt.plot(df_index.DOW, regr_index.slope * df_index.DOW + regr_index.intercept, 'r')
plt.legend(['DOW x KOSPI', regr_index_line])
plt.title(f'DOW x KOSPI (R_sqaured = {regr_index.rvalue:2f})')
plt.xlabel('DOW')
plt.ylabel('KOSPI')
plt.show();
```
